```python
#import dependencies
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np
import pandas as pd



#read csv 
animal_data = pd.read_csv("10to15 IL Shelter list.csv", sep=',', encoding='latin-1')
animal_data = animal_data.fillna(0)
animal_data
```

```python
TotalRecieved = animal_data['AllRec'].sum()
TotalAdopted = animal_data['AllAdopt'].sum()
TotalReclaim = animal_data['AllReclaim'].sum()
TotalKilled = animal_data['AllEuth'].sum()

print(TotalRecieved)
print(TotalAdopted)
print(TotalReclaim)
print(TotalKilled)
```
