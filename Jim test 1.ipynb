```python
# animal_data['YearShelterID'] = animal_data['Year'] + " " + animal_data["ShelterID"].map(str)
# animal_data.dtypes
ADGBYear = animal_data[['Year', 'AllRec', 'AllAdopt', 'AllEuth', 'AllReclaim']].copy()
ADGBYearSum = ADGBYear.groupby(['Year']).sum()
ADGBYearSum


# multiple line plot
plt.plot( 'Year', 'AllRec', data=ADGBYearSum, marker='o', markerfacecolor='blue', markersize=12, color='skyblue', linewidth=4)
plt.plot( 'Year', 'AllAdopt', data=ADGBYearSum, marker='', color='olive', linewidth=2)
plt.plot( 'Year', 'AllEuth', data=ADGBYearSum, marker='', color='olive', linewidth=2, linestyle='dashed', label="toto")
plt.legend()
```

```python
ADGBYearCount = ADGBYear.groupby(['Year']).count()
ADGBYearCount
```
